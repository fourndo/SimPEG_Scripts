```python
import dask
import dask.array as da
import numpy as np
import time
import os.path
import shutil
from dask.distributed import Client
client = Client()
client

import os
os.environ["PATH"] += os.pathsep + 'C:\\Program Files (x86)\\Graphviz2.38\\bin\\'
```

```python
tc = time.time()
nCPU = 8
M = 8000
N = 100000
maxRAM = 2 # in Gb

# Compute approximate chunks for memory limit 
rowChunk = int(np.ceil(maxRAM / (N*nCPU*8*1e-9)))
nChunks = int(np.floor(M/rowChunk))
```

```python
if os.path.exists("sens.zarr"):
  shutil.rmtree("sens.zarr")
else:
  print("The file does not exist")

def rand(rows, ind):
    return np.random.randn(rows, N)

# Construct matrix row by row
row = dask.delayed(rand, pure=True)

makeRows = [row(rowChunk,ii) for ii in  range(nChunks)]

buildMat = [da.from_delayed(makeRow, dtype=float, shape=(rowChunk, N)) for makeRow in makeRows]

stack = da.vstack(buildMat)
print(stack)
da.to_zarr(stack, "sens.zarr")
print("Write completed:" + str(time.time()-tc))

# print(np.array(F[1,:]), np.array(F[0,:]))
# vec = da.dot(F, np.ones(N))
# print(np.array(vec))
# print("Run completed:" + str(time.time()-tc))
```
