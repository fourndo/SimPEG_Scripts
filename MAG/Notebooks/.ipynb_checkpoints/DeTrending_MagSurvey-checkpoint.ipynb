```python
import numpy as np
import matplotlib.pyplot as plt
import shapefile
from SimPEG import Mesh, PF, Utils
from SimPEG.Utils import mkvc
from scipy.interpolate import griddata
from GeoToolkit.Mag import MathUtils
```

```python
# Load data
work_dir = r"C:\Users\DominiqueFournier\Dropbox\YL.YuDai(AERO)"

survey = Utils.io_utils.readUBCmagneticsObservations(work_dir + "\\Obs_Loc_Yudai_TMI_New.obs")[0]
```

```python

# Data[X,Y,Z,Mag]
A = np.c_[np.ones(survey.nD), survey.rxLoc[:,:2]]

# Compute least-squares solution for poly parameters
poly = np.linalg.solve(np.dot(A.T,A), np.dot(A.T,survey.dobs))

# Generate the first-order trend on each points
d = np.dot(A,poly)

dataDetrend =  survey.dobs - d

Utils.io_utils.writeUBCmagneticsObservations(work_dir + "\\Obs_Loc_Yudai_TMI_New_Detrended.obs", survey, dataDetrend)
```

```python
fig=plt.figure(figsize=(12,4))
ax1=plt.subplot(1,3,1)
im = Utils.plot2Ddata(survey.rxLoc, survey.dobs, ax=ax1)
plt.colorbar(im[0], orientation='horizontal')


ax2=plt.subplot(1,3,2)
im = Utils.plot2Ddata(survey.rxLoc, d, ax=ax2)
plt.colorbar(im[0], orientation='horizontal')

ax3=plt.subplot(1,3,3)
im = Utils.plot2Ddata(survey.rxLoc, dataDetrend, ax=ax3)
plt.colorbar(im[0], orientation='horizontal')
```
