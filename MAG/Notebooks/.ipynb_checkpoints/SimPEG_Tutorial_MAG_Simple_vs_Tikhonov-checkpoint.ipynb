Magnetic Vector Inversion - Demo 
==================

**Objective**

Invert mag data with MVI-S and with rotation of the objective function


```python
# Load the various libraries
from SimPEG import Mesh, Directives, Maps, InvProblem, Optimization, DataMisfit, Inversion, Utils, Regularization
import SimPEG.PF as PF
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from SimPEG.Utils import speye
```

Data
==

First we need to load some mag data.

```python
# Load some data in x,y,z,data(nT)
# dataMat = np.loadtxt('.\Synthetic\Obs_loc_TMI.obs')
# xyzLoc = dataMat[:,:3]
# data = dataMat[:,3]

# The data was simulated with a simple vertical inducing field. 
# This only matters for the data projection.
H0 = [50000, 90 , 0] # Strenght (nT), Inc (deg), Dec (deg)

# We can create a SimPEG.Survey, which contains all the necessary information needed for the inversion
xr = np.linspace(-20., 20., 20)
yr = np.linspace(-20., 20., 20)
X, Y = np.meshgrid(xr, yr)
Z = np.zeros(X.size)+1. # Let just put the observation flat

topo = np.c_[Utils.mkvc(X.T), Utils.mkvc(Y.T), Utils.mkvc(Z.T*0)]

# Create a MAGsurvey
xyzLoc = np.c_[Utils.mkvc(X.T), Utils.mkvc(Y.T), Utils.mkvc(Z.T)]
rxLoc = PF.BaseMag.RxObs(xyzLoc)
srcField = PF.BaseMag.SrcField([rxLoc], param=H0)
survey = PF.BaseMag.LinearSurvey(srcField) 

```

**Mesh**
=====
Now we need a mesh to invert physical properties on. 
I provided a mesh in the same folder, but generally you won't have one.
It is a tradeoff between having a very large problem to solve (too small cells) versus resolution (too coarse). Usually recommended to have 1 or 2 cells between each survey lines.

We created a small utility to generate a mesh based on the core cell dimension and padding distance. Use at your own risk ;)

```python
# I want 10 m cells in x, y and z covering our survey
h = np.r_[10., 10., 10.]

# Also need to paddout far enough in case there is a regional field
# Here I am sending distances N, S, E, W, above and below all at once
# Let's say 200 m padding
padDist = np.ones((3,2))*200.

mesh = Utils.modelutils.meshBuilder(xyzLoc, h, padDist,
                   meshGlobal=None,
                   expFact=1.3,
                   meshType='TREE',
                   verticalAlignment='center')

mesh = Utils.modelutils.refineTree(mesh, topo, finalize=True, dtype="surface", nCpad=[5, 3, 3])

# We can plot it to see
mesh.plotGrid()
plt.show()

# And save a copy on disk for later use
# mesh.writeUBC("MeshOctree.msh", models:{'TensorTest.msh':actv})
```

Creating the forward Problem
================

Now that we have all our spatial components, we can create our linear system. For a single location and single component of the data, the system would looks like this:

$$ b_x =
	\begin{bmatrix}
	T_{xx}^1 &... &T_{xx}^{nc} & T_{xy}^1 & ... & T_{xy}^{nc} & T_{xz}^1 & ... & T_{xz}^{nc}\\
	 \end{bmatrix}
	 \begin{bmatrix}
		\mathbf{M}_x \\ \mathbf{M}_y \\ \mathbf{M}_z
	\end{bmatrix} \\ $$

where each of $T_{xx},\;T_{xy},\;T_{xz}$ are [nc x 1] long. For the $y$ and $z$ component, we need the two other rows of the tensor $\mathbf{T}$.
The magnetization direction $\mathbf{M_x,\;M_y\;,Mz}$ are unknown parameters that we want to invert for. Inside of writing in terms of magnetization, we can write it in terms of *effective susceptibility* along cartesian directions such that:

$$ {\mathbf{d}}^{\text{pred}} = H_0 \mathbf{T} \begin{bmatrix} \kappa_x \\ \kappa_y \\ \kappa_z \end{bmatrix}$$

In most geophysical surveys, we are not collecting all three components, but rather the magnitude of the field, or $Total\;Magnetic\;Intensity$ (TMI) data.
Because the inducing field is really large, we will assume that the anomalous fields are parallel to $\vec H_0$:

$$ d^{TMI}  = \hat H_0 \cdot \vec d$$

We then end up with a much smaller system:

$$ d^{TMI} = \mathbf{F\; \kappa_e}$$

where $\mathbf{F} \in \mathbb{R}^{nd \times nc}$ is our $forward$ operator.

```python
# First we need to discretize the topo onto the mesh
# This function will assign 0:air, 1:ground
actv = Utils.modelutils.surface2ind_topo(mesh, topo, gridLoc='CC', method='nearest')
nC = int(np.sum(actv))

# Creat reduced identity map
idenMap = Maps.IdentityMap(nP=3*nC)

# Create the forward model operator
prob = PF.Magnetics.MagneticIntegral(mesh, chiMap=idenMap,
                                     actInd=actv, modelType='vector',
                                    parallelized=True)
# Pair the survey and problem
survey.pair(prob)
```

```python
# Add a block 
ind = Utils.ModelBuilder.getIndicesBlock(np.r_[-10,-10,-30], np.r_[10,10,-10], mesh.gridCC)
model = np.zeros(mesh.nC)
model[ind] = 0.01

actvMap = Maps.InjectActiveCells(mesh, actv, 0)
mesh.writeUBC('OctreeMesh.msh', models={"Model.sus":model})

dobs = prob.fields(np.r_[np.zeros(actv.sum()*2),model[actv] ])
survey.dobs = dobs
survey.std = np.ones_like(dobs) * np.abs(dobs).min()


# We can check how the discretized topography looks
plt.figure()
ax = plt.subplot()
mesh.plotSlice(actv, ax = ax, normal = 'Y', grid=True)
ax.set_title('EW-Section of topography')
ax.set_xlabel('x');ax.set_ylabel('y')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Looking good
# We can plot the data, looks like this
# We can now generate data
im = Utils.PlotUtils.plot2Ddata(survey.srcField.rxList[0].locs, survey.dobs)
plt.colorbar(im[0])
```

Once we have our problem, we can use the inversion tools in SimPEG to run our inversion:

```python
theta = np.ones(mesh.nC)*np.deg2rad(0.)
phi = np.ones(mesh.nC)*np.deg2rad(45.)


indActive = np.zeros(mesh.nC, dtype=bool)
indActive[actv] = True

# This is for the aircells
Pac = Utils.speye(mesh.nC)[:, indActive]

Dx1 = Pac.T * Regularization.getDiffOpRot(mesh, np.deg2rad(0.), theta, phi, 'X') * Pac
Dy1 = Pac.T * Regularization.getDiffOpRot(mesh, np.deg2rad(0.), theta, phi, 'Y') * Pac
Dz1 = Pac.T * Regularization.getDiffOpRot(mesh, np.deg2rad(0.), theta, phi, 'Z') * Pac

Dx2 = Pac.T * Regularization.getDiffOpRot(mesh, np.deg2rad(0.), theta, phi, 'X', forward=False) * Pac
Dy2 = Pac.T * Regularization.getDiffOpRot(mesh, np.deg2rad(0.), theta, phi, 'Y', forward=False) * Pac
Dz2 = Pac.T * Regularization.getDiffOpRot(mesh, np.deg2rad(0.), theta, phi, 'Z', forward=False) * Pac
```

```python

# This Mapping connects all the regularizations together
wires = Maps.Wires(('p', nC), ('s', nC), ('t', nC))


mstart = np.ones(3*nC) * 1e-4
mref = np.zeros(3*nC)

prob.model = mstart

dpre0 = prob.fields(mstart)
wr = prob.getJtJdiag(mstart)**0.5
wr = (wr/np.max(wr))

# Create three regularization for the different components
# of magnetization
reg_p = Regularization.Sparse(mesh, indActive=actv, mapping=wires.p)
reg_p.cell_weights = (wires.p * wr)
reg_p.mref = mref

reg_s = Regularization.Sparse(mesh, indActive=actv, mapping=wires.s)
reg_s.cell_weights = (wires.s * wr)
reg_s.mref = mref

reg_t = Regularization.Sparse(mesh, indActive=actv, mapping=wires.t)
reg_t.cell_weights = (wires.t * wr)
reg_t.mref = mref

reg = reg_p + reg_s + reg_t
reg.mref = mref

# Data misfit function
dmis = DataMisfit.l2_DataMisfit(survey)
dmis.W = 1./survey.std

# Choose a solver
opt = Optimization.ProjectedGNCG(maxIter=20, lower=-10., upper=10.,
                                 maxIterCG=20, tolCG=1e-3)

# The inverse problem needs to how the misfit, regularizer and solver
invProb = InvProblem.BaseInvProblem(dmis, reg, opt)

# Add directives to the inversion
betaest = Directives.BetaEstimate_ByEig()

# Here is where the norms are applied usually
IRLS = Directives.Update_IRLS(f_min_change=1e-3,
                              minGNiter=1)

update_Jacobi = Directives.UpdatePreconditioner()
targetMisfit = Directives.TargetMisfit()
# MVIS = Directives.MagneticVectorSparse()

# Create active map to go from reduce set to full
actvMap = Maps.InjectActiveCells(mesh, actv, 0)
saveModel = Directives.SaveUBCModelEveryIteration(mapping=actvMap, vector=True)
saveModel.fileName = 'MVI_C'

# Connect all the pieces together
inv = Inversion.BaseInversion(invProb,
                              directiveList=[betaest, IRLS, update_Jacobi,
                                            saveModel])

# Invert with a starting model
mrec_MVI = inv.run(mstart)
```

```python
beta = invProb.beta

# %% RUN MVI-S WITH SPARSITY

# # STEP 3: Finish inversion with spherical formulation
mstart = Utils.matutils.xyz2atp(mrec_MVI.reshape((nC,3),order='F'))

# for misfit in ComboMisfit.objfcts:
prob.coordinate_system = 'spherical'
prob.model = mstart
dd = prob.fields(mstart)
# Create a block diagonal regularization
wires = Maps.Wires(('amp', nC), ('theta', nC), ('phi', nC))

# Create a regularization x3 x2 (Forward + Backward)
reg_a = Regularization.Sparse(mesh, indActive=actv,
                              mapping=wires.amp, gradientType='component')
reg_a.norms = np.c_[0,0,0,0]
reg_a.mref = mref
reg_a.objfcts[1].regmesh._cellDiffxStencil = Dx1
reg_a.objfcts[1].regmesh._aveCC2Fx = speye(nC)
reg_a.objfcts[2].regmesh._cellDiffyStencil = Dy1
reg_a.objfcts[2].regmesh._aveCC2Fy = speye(nC)
reg_a.objfcts[3].regmesh._cellDiffzStencil = Dz1
reg_a.objfcts[3].regmesh._aveCC2Fz = speye(nC)

reg_t = Regularization.Sparse(mesh, indActive=actv,
                              mapping=wires.theta, gradientType='component')
reg_t.alpha_s = 0.  # No reference angle
reg_t.space = 'spherical'
reg_t.norms = np.c_[0,0,0,0]
reg_t.eps_q = 5e-2
reg_t.mref = mref
reg_t.objfcts[1].regmesh._cellDiffxStencil = Dx1
reg_t.objfcts[1].regmesh._aveCC2Fx = speye(nC)
reg_t.objfcts[2].regmesh._cellDiffyStencil = Dy1
reg_t.objfcts[2].regmesh._aveCC2Fy = speye(nC)
reg_t.objfcts[3].regmesh._cellDiffzStencil = Dz1
reg_t.objfcts[3].regmesh._aveCC2Fz = speye(nC)
# reg_t.alpha_x, reg_t.alpha_y, reg_t.alpha_z = 0.25, 0.25, 0.25

reg_p = Regularization.Sparse(mesh, indActive=actv,
                              mapping=wires.phi, gradientType='component')
reg_p.alpha_s = 0.  # No reference angle
reg_p.space = 'spherical'
reg_p.norms = np.c_[0,0,0,0]
reg_p.eps_q = 5e-2
reg_p.mref = mref
reg_p.objfcts[1].regmesh._cellDiffxStencil = Dx1
reg_p.objfcts[1].regmesh._aveCC2Fx = speye(nC)
reg_p.objfcts[2].regmesh._cellDiffyStencil = Dy1
reg_p.objfcts[2].regmesh._aveCC2Fy = speye(nC)
reg_p.objfcts[3].regmesh._cellDiffzStencil = Dz1
reg_p.objfcts[3].regmesh._aveCC2Fz = speye(nC)

# Create a regularization x3 x2 (Forward + Backward)
reg_a_bck = Regularization.Sparse(mesh, indActive=actv,
                              mapping=wires.amp, gradientType='component')
reg_a_bck.norms = np.c_[0,0,0,0]
reg_a_bck.mref = mref
reg_a_bck.objfcts[1].regmesh._cellDiffxStencil = Dx2
reg_a_bck.objfcts[1].regmesh._aveCC2Fx = speye(nC)
reg_a_bck.objfcts[2].regmesh._cellDiffyStencil = Dy2
reg_a_bck.objfcts[2].regmesh._aveCC2Fy = speye(nC)
reg_a_bck.objfcts[3].regmesh._cellDiffzStencil = Dz2
reg_a_bck.objfcts[3].regmesh._aveCC2Fz = speye(nC)

reg_t_bck = Regularization.Sparse(mesh, indActive=actv,
                              mapping=wires.theta, gradientType='component')
reg_t_bck.alpha_s = 0.  # No reference angle
reg_t_bck.space = 'spherical'
reg_t_bck.norms = np.c_[0,0,0,0]
reg_t_bck.eps_q = 5e-2
reg_t_bck.mref = mref
reg_t_bck.objfcts[1].regmesh._cellDiffxStencil = Dx2
reg_t_bck.objfcts[1].regmesh._aveCC2Fx = speye(nC)
reg_t_bck.objfcts[2].regmesh._cellDiffyStencil = Dy2
reg_t_bck.objfcts[2].regmesh._aveCC2Fy = speye(nC)
reg_t_bck.objfcts[3].regmesh._cellDiffzStencil = Dz2
reg_t_bck.objfcts[3].regmesh._aveCC2Fz = speye(nC)
# reg_t.alpha_x, reg_t.alpha_y, reg_t.alpha_z = 0.25, 0.25, 0.25

reg_p_bck = Regularization.Sparse(mesh, indActive=actv,
                              mapping=wires.phi, gradientType='component')
reg_p_bck.alpha_s = 0.  # No reference angle
reg_p_bck.space = 'spherical'
reg_p_bck.norms = np.c_[0,0,0,0]
reg_p_bck.eps_q = 5e-2
reg_p_bck.mref = mref
reg_p_bck.objfcts[1].regmesh._cellDiffxStencil = Dx2
reg_p_bck.objfcts[1].regmesh._aveCC2Fx = speye(nC)
reg_p_bck.objfcts[2].regmesh._cellDiffyStencil = Dy2
reg_p_bck.objfcts[2].regmesh._aveCC2Fy = speye(nC)
reg_p_bck.objfcts[3].regmesh._cellDiffzStencil = Dz2
reg_p_bck.objfcts[3].regmesh._aveCC2Fz = speye(nC)

reg = reg_a + reg_t + reg_p + reg_a_bck + reg_t_bck + reg_p_bck
reg.mref = mref

Lbound = np.kron(np.asarray([0, -np.inf, -np.inf]), np.ones(nC))
Ubound = np.kron(np.asarray([10, np.inf, np.inf]), np.ones(nC))

# Data misfit function
dmis = DataMisfit.l2_DataMisfit(survey)
dmis.W = 1./survey.std

# Add directives to the inversion
opt = Optimization.ProjectedGNCG(maxIter=40,
                                 lower=Lbound,
                                 upper=Ubound,
                                 maxIterLS=10,
                                 maxIterCG=20, tolCG=1e-3,
                                 stepOffBoundsFact=1e-8)

invProb = InvProblem.BaseInvProblem(dmis, reg, opt, beta=beta)
#  betaest = Directives.BetaEstimate_ByEig()

# Here is where the norms are applied
IRLS = Directives.Update_IRLS(f_min_change=1e-4, maxIRLSiter=40,
                              minGNiter=1, beta_tol = 0.5, prctile=100,
                              coolingRate=1, coolEps_q=True,
                              betaSearch=True)

# Special directive specific to the mag amplitude problem. The sensitivity
# weights are update between each iteration.
ProjSpherical = Directives.ProjSpherical()
update_SensWeight = Directives.UpdateSensitivityWeights()
update_Jacobi = Directives.UpdatePreconditioner()
saveModel = Directives.SaveUBCModelEveryIteration(mapping=actvMap, vector=True)
saveModel.fileName = 'MVI_S'

inv = Inversion.BaseInversion(invProb,
                              directiveList=[ProjSpherical, IRLS, update_SensWeight,
                                             update_Jacobi, saveModel])

mrec_MVI_S = inv.run(mstart)
```

```python
IRLS.reg.objfcts[0].objfcts
```

```python
# Here is a quick way to convert your vector into a scalar
amp = np.sum(mrec_MVI.reshape((nC,3), order='F')**2., axis=1)**0.5

# We also need to map the model from ground-only to full mesh
actvMap = Maps.InjectActiveCells(mesh, actv, 0)
amp = actvMap * amp

# Et voila, now we can just slice it by changing the ind=#
plt.figure()
ax = plt.subplot()
im = mesh.plotSlice(amp, ax = ax, normal = 'Y', ind=20)
ax.set_title('Recovered magnetic amplitude')
ax.set_xlabel('x');ax.set_ylabel('y')
plt.colorbar(im[0])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

Predicted data
========

Always very important after an inversion is to look at the data residual.

We want to make sure the inversion can actually predict the observed data, and maybe adjust the uncertainties if not satisfactory.

```python
fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(2,2,1)
Utils.PlotUtils.plot2Ddata(survey.srcField.rxList[0].locs, survey.dobs, ax= ax1)

# Predicted
ax2 = plt.subplot(2,2,2)
Utils.PlotUtils.plot2Ddata(survey.srcField.rxList[0].locs, invProb.dpred, ax= ax2)

# Residual
ax3 = plt.subplot(2,2,4)
Utils.PlotUtils.plot2Ddata(survey.srcField.rxList[0].locs, survey.dobs-invProb.dpred, ax= ax3)

plt.show()
```
