```python
import numpy as np
import matplotlib.pyplot as plt
import shapefile
from SimPEG import Mesh, PF, Utils
from SimPEG.Utils import mkvc
from scipy.interpolate import griddata
from GeoToolkit.Mag import MathUtils
```

```python
workDir = ".\\"
dataDir = "..\\Data\\"



world = shapefile.Reader(dataDir + "conjunto_de_datos\\curva_nivel50_l.shp")
# Extract lines from shape file
X, Y = [], []
Z = []
for shape in world.shapeRecords():

    for ii, part in enumerate(shape.shape.parts):

        if ii != len(shape.shape.parts)-1:
            x = [i[0] for i in shape.shape.points[shape.shape.parts[ii]:shape.shape.parts[ii+1]]]
            y = [i[1] for i in shape.shape.points[shape.shape.parts[ii]:shape.shape.parts[ii+1]]]

        else:
            x = [i[0] for i in shape.shape.points[shape.shape.parts[ii]:]]
            y = [i[1] for i in shape.shape.points[shape.shape.parts[ii]:]]


        X.append(np.vstack(x))
        Y.append(np.vstack(y))
        Z.append(np.ones_like(np.vstack(y)) * shape.record["ELEVACION"])
                

```

```python
# Interpolate topograpgy on regular grid
xx, yy = np.arange(np.vstack(X).min(), np.vstack(X).max(), 150), np.arange(np.vstack(Y).min(), np.vstack(Y).max(), 150)

xGrid, yGrid = np.meshgrid(xx, yy)
zGrid = np.squeeze(griddata(np.c_[np.vstack(X),np.vstack(Y)], np.vstack(Z), (xGrid, yGrid), method='nearest'))

ax = plt.subplot()
ax.imshow(zGrid)
topo = np.c_[mkvc(xGrid), mkvc(yGrid), mkvc(zGrid)]

# Write to file
np.savetxt(workDir + "Topo.dat", topo)
```

```python
# Import mag data
data = np.loadtxt(dataDir + "g12b25 Magnetic data Baroyeca.txt", skiprows=8)

_, zInterp = MathUtils.minCurvatureInterp(topo[:,:2], topo[:,2], 
                                          xyzOut = data[:,:2])

ind = ~np.isnan(zInterp)
Utils.plot2Ddata(data[ind,:2], zInterp[ind])
```

```python
# Data[X,Y,Z,Mag]
A = np.c_[np.ones(ind.sum()), data[ind,:2]]

# Compute least-squares solution for poly parameters
poly = np.linalg.solve(np.dot(A.T,A), np.dot(A.T,data[ind,3]))

# Generate the first-order trend on each points
d = np.dot(A,poly)

dataDetrend =  data[ind,3] - d
```

```python
Utils.plot2Ddata(data[ind,:2], data[ind,3]-d)
```

```python
# Create a mag survey and write to file
rxLoc = PF.BaseMag.RxObs(np.c_[data[ind,:2], zInterp[ind] + 300])
srcField = PF.BaseMag.SrcField([rxLoc], param=[47515, 55, 10+49/60])
survey = PF.BaseMag.LinearSurvey(srcField)
survey.dobs = data[ind, 3]
survey.std = np.ones_like(data[ind, 3])

Utils.io_utils.writeUBCmagneticsObservations(workDir + "Baroyeca.mag", survey, survey.dobs)
```

```python
mesh = Mesh.TensorMesh([np.ones_like(xx)*150, np.ones_like(yy)*150, np.ones(20)*50], x0=[xx.min(0), yy.min(), zInterp[ind].min()-1000])
mesh.writeUBC(workDir + "Mesh100.msh")
```
