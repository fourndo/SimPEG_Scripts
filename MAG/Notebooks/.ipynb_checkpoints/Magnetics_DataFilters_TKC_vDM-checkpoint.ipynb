**Where are the diamonds: Using Earth's Potential Fields (Magnetics)** 

In this tutorial we will run a magnetic inverse problem from an input file.
The synthetic example used is based on the TKC kimberlite deposit, Northwest Territories. We are attempting to use the magnetic field to image the various rock units making up the diamond deposit.
We will skip the model creation for now, at it is quite involved and out of scope for this notebook. 

```python
## First we need to load all the libraries and set up the path
## for the input files. Same files as used by the online tutorial
## Need to be on \pf\dev branch !!!
%pylab inline 
import SimPEG.PF as PF
from SimPEG import *
from SimPEG.Utils import io_utils
import matplotlib
import matplotlib.colors as colors
import scipy as sp
import time as tm
import os
import urllib
import shutil
from ipywidgets.widgets import interact, IntSlider

# URL and files needed for this notebook
#url='https://storage.googleapis.com/simpeg/tests/potential_fields/'
#cloudfiles = ['GravData.obs','Gaussian.topo','Mesh_10m.msh','ModelStart.sus','Obs_loc_TMI.obs','PYMAG3D_inv.inp']
url='https://storage.googleapis.com/simpeg/tkc_synthetic/potential_fields/'
cloudfiles = ['MagData.obs']
# Create a temporary directory with all files downloaded from cloud
# for file in cloudfiles:
io_utils.download(url+cloudfiles[0], overwrite=True)
```

```python
# Load the data and plot
survey = PF.Magnetics.readMagneticsObservations(cloudfiles[0])

rxLoc = survey.srcField.rxList[0].locs
PF.Magnetics.plot_obs_2D(rxLoc,survey.dobs)
```

**Setup**

We want to filter the data and extract structural information.  


```python
from scipy.interpolate import griddata

# First, grid the data on a pre-defined resolution
# Will do something simple here and just fix it to 50
npts = 50

# Create grid of points
x = np.linspace(rxLoc[:, 0].min(), rxLoc[:, 0].max(), npts)
y = np.linspace(rxLoc[:, 1].min(), rxLoc[:, 1].max(), npts)

X, Y = np.meshgrid(x, y)

# Interpolate the data on the grid
d_grid = griddata(rxLoc[:, 0:2], survey.dobs, (X, Y), method='cubic')

# We can plot the data
plt.imshow(d_grid, extent=[x.min(), x.max(), y.min(), y.max()],
                   origin='lower')
plt.colorbar()
```

```python
# Figure out padding width
padx = int(np.floor(d_grid.shape[1]))
pady = int(np.floor(d_grid.shape[0]))

# Add paddings
dpad = np.c_[np.fliplr(d_grid[:,0:padx]),
             d_grid,
             np.fliplr(d_grid[:,-padx:])
            ]

dpad = np.r_[np.flipud(dpad[0:pady,:]),
             dpad,
             np.flipud(dpad[-pady:,:])
            ]

dx = (1)/(dpad.shape[1]-1)
dy = (1)/(dpad.shape[0]-1)

kx = np.fft.fftfreq(dpad.shape[1], dx)
ky = np.fft.fftfreq(dpad.shape[0], dy)

Kx, Ky = numpy.meshgrid(ky, kx)

plt.figure(figsize=(14,4))
plt.subplot(1,3,1)
plt.imshow(dpad,origin='lower')

# Create a position array
[U,V] = meshgrid(np.linspace(np.floor(-dpad.shape[0]/2),np.ceil(dpad.shape[0]/2),dpad.shape[0]),
                 np.linspace(np.floor(-dpad.shape[1]/2),np.ceil(dpad.shape[1]/2),dpad.shape[1]))


# Tapper the paddings
rampx = -np.cos(3*np.pi*np.asarray(range(int(dpad.shape[0]/3)))/dpad.shape[0])
rampx = np.r_[rampx,np.ones(int(dpad.shape[0]/3)),-rampx]/2. + 0.5
tapperx,_ = meshgrid(rampx,np.ones(dpad.shape[1]))
# tapperx[padx:-padx,:] = 1.

rampy = -np.cos(3*np.pi*np.asarray(range(int(dpad.shape[1]/3)))/dpad.shape[1])
rampy = np.r_[rampy,np.ones(int(dpad.shape[0]/3)),-rampy]/2. +0.5
_, tappery = meshgrid(np.ones(dpad.shape[0]), rampy)

dpad_taped = tapperx*tappery*dpad 
plt.subplot(1,3,2)
plt.imshow(tappery*tapperx,origin='lower')

plt.subplot(1,3,3)
plt.imshow(dpad_taped,origin='lower')
```

```python
# Now that we have a grid we can run the FFT
FS = np.fft.fft2(dpad_taped)

# And bring it back to the data space
IFS = np.fft.ifft2(FS)

# Extract core
d_grid_out = np.real(IFS[pady:-pady, padx:-padx])

# Lets view the FFT, inversed FFT and residual maps
plt.figure(figsize=(12,12))
ax1=plt.subplot(2,2,1)
plt.imshow(np.log(np.abs(np.fft.fftshift(FS))**2))
ax1.set_title('2D FFT')
ax2=plt.subplot(2,2,2)
im=plt.imshow(d_grid, extent=[x.min(), x.max(), y.min(), y.max()],
                   origin='lower')
plt.colorbar(im)

ax2=plt.subplot(2,2,3)
im=plt.imshow(d_grid_out, extent=[x.min(), x.max(), y.min(), y.max()],
                   origin='lower')
plt.colorbar(im)
ax2.set_title('Inverse FFT')
ax3=plt.subplot(2,2,4)
im = plt.imshow(d_grid_out-d_grid, extent=[x.min(), x.max(), y.min(), y.max()],
                   origin='lower')
plt.colorbar(im)
ax3.set_title('Residual')
```

```python
# Calculate First order derivative
m, n = 1, 1

FHxD = (Kx*1j)*FS
fhxd_pad = np.fft.ifft2(FHxD)
fhxd = np.real(fhxd_pad[pady:-pady, padx:-padx])

FHyD = (Ky*1j)*FS
fhyd_pad = np.fft.ifft2(FHyD)
fhyd = np.real(fhyd_pad[pady:-pady, padx:-padx])

FHzD = FS*np.sqrt(Kx**2 + Ky**2)
fhzd_pad = np.fft.ifft2(FHzD)
fhzd = np.real(fhzd_pad[pady:-pady, padx:-padx])


horiz_deriv = np.sqrt(fhxd**2 + fhyd**2)
tilt = np.arctan2(fhzd, horiz_deriv)
    
# Lets view the FFT, inversed FFT and residual maps
plt.figure(figsize=(10,10))
ax2=plt.subplot(3,3,1)
plt.imshow(d_grid, extent=[x.min(), x.max(), y.min(), y.max()],
                   origin='lower')
ax2.set_title('TMI - Data')
plt.colorbar()

ax2=plt.subplot(3,3,2)
im=plt.imshow(fhxd,origin='lower')
plt.colorbar(im)
ax2.set_title('Horizontal Derivative - X')

ax2=plt.subplot(3,3,3)
im=plt.imshow(fhyd,origin='lower')
plt.colorbar(im)
ax2.set_title('Horizontal Derivative - Y')

ax2=plt.subplot(3,3,4)
im=plt.imshow(fhzd,origin='lower')
plt.colorbar(im)
ax2.set_title('Vertical Derivative - Y')

ax2=plt.subplot(3,3,5)
im=plt.imshow(horiz_deriv,origin='lower')
plt.colorbar(im)
ax2.set_title('Horizontal Total')

ax2=plt.subplot(3,3,6)
im=plt.imshow(tilt,origin='lower')
plt.colorbar(im)
ax2.set_title('Tilt')
```

```python
# Repeat the filters with current version of Fatiando 
from fatiando.gravmag import transform

fhdx = transform.derivx(mkvc(X), mkvc(Y), mkvc(d_grid), d_grid.shape, method='fft')
fhdy = transform.derivy(mkvc(X), mkvc(Y), mkvc(d_grid), d_grid.shape, method='fft')
```

```python
# Lets view the FFT, inversed FFT and residual maps
plt.figure(figsize=(15,8))

ax2=plt.subplot(1,2,1)
im=plt.imshow(fhdx.reshape(d_grid.shape, order='F'),origin='lower')
plt.colorbar(im)
ax2.set_title('Horizontal Derivative - X')

ax2=plt.subplot(1,2,2)
im=plt.imshow(fhdy.reshape(d_grid.shape, order='F'),origin='lower')
plt.colorbar(im)
ax2.set_title('Horizontal Derivative - Y')
```
